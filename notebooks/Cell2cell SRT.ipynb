```python
import pandas
import mba263
import matplotlib.pyplot as plt
```

```python
data = pandas.read_csv('cell2cell.csv')
```

```python
data.head(10)
```

```python
data.describe()
```

Creating separate datasets for calibration and validation

```python
calibrate = (data[data['calibrat']==1]).copy()
validate = (data[data['calibrat']==0]).copy()
```

```python
validate.describe()
```

```python
calibrate.head(10)
```

```python
calibrate['churn'].describe()
```

```python
validate['churn'].describe()
```

```python
varlist =[ 
'revenue',
'mou',
'recchrge',
'directas',
'overage',
'roam',
'changem',
'changer',
'dropvce',
'blckvce',
'unansvce',
'custcare',
'threeway',
'mourec',
'outcalls',
'incalls',
'peakvce',
'opeakvce',
'dropblk',
'callfwdv',
'callwait',
'months',
'uniqsubs',
'actvsubs',
'phones',
'models',
'eqpdays',
'age1',
'age2',
'children',
'credita',
'creditaa',
'prizmrur',
'prizmub',
'prizmtwn',
'refurb',
'webcap',
'truck',
'rv',
'occprof',
'occcler',
'occcrft',
'occstud',
'occhmkr',
'occret',
'occself',
'ownrent',
'marryun',
'marryyes',
'mailord',
'mailres',
'mailflag',
'travel',
'pcown',
'creditcd',
'retcalls',
'retaccpt',
'newcelly',
'newcelln',
'refer',
'incmiss',
'income',
'mcycle',
'setprcm',
'setprc',
'retcall'
]
```

```python
result=mba263.logit(calibrate['churndep'],calibrate[varlist])
```

```python
pandas.set_option('display.max_rows', None)
mba263.odds_ratios(result)
```

```python
validate['p_logit']=result.predict(validate[varlist])
```

```python
lift_logit = mba263.lift(validate['churn'],validate['p_logit'])
lift_logit
```

```python
plt.plot(lift_logit)
```
